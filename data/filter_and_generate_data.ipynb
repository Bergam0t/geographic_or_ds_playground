## Demand

```python
population_df = pd.read_excel("sapelsoasyoa20222024.xlsx", sheet_name="Mid-2024 LSOA 2021", skiprows=3)
```

```python
population_df.head()
```

```python
[i for i in population_df.columns]
```

```python
population_df_filtered = population_df[population_df["LSOA 2021 Name"].str.contains(
    "Devon|Torbay|South Hams|Exeter|Torridge|Teignbridge|Plymouth"
    )]

population_df_filtered.head()
```

```python
population_df_filtered["F50-84"] = population_df_filtered[
    [f"F{i}" for i in range(50, 84)]
].sum(axis=1)

population_df_filtered["M50-84"] = population_df_filtered[
    [f"M{i}" for i in range(50, 84)]
].sum(axis=1)

# population_df_filtered["F18-49"] = population_df_filtered[
#     [f"F{i}" for i in range(18, 49)]
# ].sum(axis=1)

# population_df_filtered["M18-49"] = population_df_filtered[
#     [f"M{i}" for i in range(18, 49)]
# ].sum(axis=1)

# population_df_filtered["F71+"] = population_df_filtered[
#     [f"F{i}" for i in range(71, 90)]
# ].sum(axis=1)

# population_df_filtered["M71+"] = population_df_filtered[
#     [f"M{i}" for i in range(71, 90)]
# ].sum(axis=1)
```

```python
# demand_df = population_df_filtered[["LSOA 2021 Code", "LSOA 2021 Name", "Total", "F18-49", "M18-49", "F50-70", "M50-70", "F71+", "M71+"]].copy()

demand_df = population_df_filtered[["LSOA 2021 Code", "LSOA 2021 Name", "Total", "F50-84", "M50-84"]].copy()
demand_df["MF50-84"] = demand_df["F50-84"] + demand_df["M50-84"]
```
